```python
import os
import json
import datetime as dt
import numpy as np
from osgeo import gdal

from geo_utils import ChipExtents
from json_utils import JSONReader

WKT = '''PROJCS["Albers",GEOGCS["WGS 84",DATUM["WGS_1984",SPHEROID["WGS 84",6378140,298.2569999999957,AUTHORITY["EPSG",
"7030"]],AUTHORITY["EPSG","6326"]],PRIMEM["Greenwich",0],UNIT["degree",0.0174532925199433],AUTHORITY["EPSG","4326"]],
PROJECTION["Albers_Conic_Equal_Area"],PARAMETER["standard_parallel_1",29.5],PARAMETER["standard_parallel_2",45.5],
PARAMETER["latitude_of_center",23],PARAMETER["longitude_of_center",-96],PARAMETER["false_easting",0],
PARAMETER["false_northing",0],UNIT["metre",1,AUTHORITY["EPSG","9001"]]]'''
```

```python
H = 27
V = 8

json_dir = r"Z:\sites\chesapeake\pyccd-results\H27V08\2017.08.18\json"
out_dir = r"C:\Users\dzelenak\Workspace\h27v07\time_segments\test"

if not os.path.exists(out_dir):
    
    os.makedirs(out_dir)
    
chip_info = ChipExtents(H, V)
chip_extents = chip_info.CHIP_EXTENTS
```

```python
json_reader = JSONReader(json_dir)
```

```python
chip = 315
mask = np.zeros(shape=(100,100), dtype=np.bool).flatten()
```

```python
# Create a 1-dimentional numpy array of dictionaries that hold the json information for each pixel in the chip

json_results = json_reader.get_jsonchip(h=H, v=V, chip_coord=chip_extents[chip]).flatten()
```

```python
for index, result in enumerate(json_results):
    
    if len(result["change_models"]) > 0:
        
        mask[index] = json_reader.check_time_segment(results=result)
        
    else:
        
        # case for when PyCCD coverage = 0
        
        mask[index] = 0
        
out_mask = mask.reshape((100,100))
```

```python
driver = gdal.GetDriverByName("GTiff")

out_str = out_dir + os.sep + "chips" + os.sep + "mask_" + str(chip) + ".tif"

if not os.path.exists(os.path.split(out_str)[0]):
    
    os.makedirs(os.path.split(out_str)[0])
    
outfile = driver.Create(out_str, 100, 100, 1, gdal.GDT_Byte)

outband = outfile.GetRasterBand(1)

outband.WriteArray(out_mask)

outband.FlushCache()

geo_transform = (chip_extents[chip].x_min, 30.0, 0.0, chip_extents[chip].y_max, 0.0, -30.0)



outfile.SetGeoTransform(geo_transform)

outfile.SetProjection(WKT)

outfile = None
```
